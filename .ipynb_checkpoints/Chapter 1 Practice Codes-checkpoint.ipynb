# Analog Vs Digital Signals

What is a Signal?

A signal is a way to represent information that changes over time or space.<br>
Think of it as a message or data that is transmitted, processed, or stored. <br>
Signals are all around us and are used to describe things like sound, light, temperature, or even motion.

- A signal is like a pattern or variation that carries information.
- It can be something you hear, see, or measure.
- Signals can be processed to extract useful information, like understanding a phone call or analyzing a heart rate

- Examples of Signals:

    -Sound from a Speaker: When you listen to music, the sound wave is a signal that carries the song’s information (like melody and lyrics).
    
    -Temperature Over Time: A weather station recording the temperature every hour creates a signal that shows how temperature changes throughout the day.
    
    -Video Feed: A video from your phone is a signal that carries visual information, changing frame by frame to show motion.
    
    -Heart Rate Monitor: The electrical activity from your heartbeat, measured by an ECG machine, is a signal showing your heart’s rhythm.

- ***Analog Signals:***
            
    - An analog signal is a continuous signal that varies smoothly over time.<br> 
    - It can take any value within a range and is usually a direct representation of the real-world phenomenon.

- ***Characteristics:***

    - Continuous in both time and value (no gaps or jumps).
    - Can have infinite possible values within a range.
    - Often found in nature or older technologies.

- ***Digital Signals***

    - A digital signal is a discrete signal, represented by a sequence of specific values (usually 0s and 1s) at distinct points in time. <br> 
    - It’s like breaking a continuous signal into small, separate pieces

- ***Characterstics:***

    - Discrete in time (values are taken at specific intervals, like snapshots).
    - Discrete in value (only certain values are allowed, often binary: 0 or 1).
    - Used in modern computers, smartphones, and digital devices.

Let us now look at example visualization of analog and digital signal using Python

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#analog signal 

t_cont = np.linspace(0, 1, 10000)
f= 5
A=1
analog_sig = A*np.sin(2*np.pi*f*t_cont)

fs = 100

t_disc = np.linspace(0, 1, fs, endpoint=False)


digit_signal = A*np.sin(2*np.pi*f*t_disc)

n_level = 16 

quantized_signal = np.round(((digit_signal+1)/2)*(n_level-1))

quantized_signal = (quantized_signal / (n_level-1))*2-1

plt.figure(figsize=(10, 5))
plt.plot(t_cont, analog_signal, label='Analgo Signal', color='blue')
plt.stem(t_disc, quantized_signal, linefmt='r-', markerfmt='ro', basefmt='k-', label='Digital Signal')
plt.title('Analog Vs Digital Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
